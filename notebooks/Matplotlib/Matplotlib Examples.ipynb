```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
values = [0,1,2,3,4,5,6,7]
squares = [value ** 2 for value in values] # 'value ** 2' ridicam la putere a 2
plt.plot(values, squares)
plt.show()
```

```python
plt.scatter(values, squares)
plt.show()
```

```python
values2 = list(range(100))
pows = [pow(value,3) for value in values2]
plt.scatter(values2, pows)
plt.show()
```

```python
plt.scatter(values2, pows, c=pows, cmap=plt.cm.Blues, edgecolor='none', s=10)
plt.show()
```

```python
values = list(range(50))
squares = [pow(value,2) for value in values]
cubes = [pow(value,3) for value in values]

fig, (ax1,ax2) = plt.subplots(2,1, sharex=True) # axa x o sa fie comuna
ax1.scatter(values, squares)
ax1.set_title('Squares Numbers')
ax2.plot(values, cubes)
ax2.set_title('Cubes NUmbers')
plt.show()
```

```python
values = list(range(10))
squares = [pow(value,2) for value in values]
cubes = [pow(value,3) for value in values]

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True) # axa y o sa fie comuna
ax1.scatter(values, squares)
ax1.set_title('Squares Numbers')
ax2.plot(values, cubes)
ax2.set_title('Cubes NUmbers')
plt.show()
```

```python
data= {
    'Year':[2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 150, 134, 211, 400, 522]
}

df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='line')
plt.title('Yearly Sales')
plt.show()
```

```python
data = {
    'Category': ['FIRST', 'FIRST','SECOND', 'SECOND', 'THIRD', 'THIRD','SECOND'],
    'Value': [10,20,25,45,45,50,60]
}

df = pd.DataFrame(data)
df
```

```python
df.boxplot(by='Category', column='Value')
plt.title('Values distribution by category')
plt.show()
```

```python
data = {
    'Month':['Jan','Feb','Mar','Apr','May'],
    'Week 1 Temp': [3,4,5,6,7],
    'Week 2 Temp': [8,9,10,11,12],
    'Week 3 Temp': [12,12,13,14,15],
    'Week 4 Temp': [15,16,17,18,20]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.set_index('Month', inplace=True) 
df.head()
```

```python
sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title('Weekly temperature data')
plt.show()
```
