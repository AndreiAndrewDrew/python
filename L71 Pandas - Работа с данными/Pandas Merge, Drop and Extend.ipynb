```python
import pandas as pd 
import numpy as np
```

```python
forum_users = {
    'User ID':np.array([1, 2, 3, 4, 5]), # Putem sa folosim masive numpay pnetru creaea DataFrame
    'Username':['andrew_b', 'ion_a', 'jane_smith', 'alex123', 'bob56'],
    'Age':[34, 53, 24, 16, 65],
    'Joined Date':pd.to_datetime(['2021-01-10', '2022-03-14', '2020-01-24', '2023-03-26', '2018-11-25']),
    'Total Posts':[150, 300, 250, 30, 90],
    'Reputation':[500, 720, 600, 300, 980]
}

# Creem DataFrames 

df = pd.DataFrame(forum_users)
df
```

## Add column

```python
# Adaugam o coloana noua
df['Language'] = 'English'
df
```

```python
# Coloana noua cu valori
df['Active'] = [True, False, False, True, True]
df
```

## Concatenare a doua DataFame

```python
# Creem DataFrame nou 'df2'
total_comments = {
    'Total Comments':[70, 20, 44 ,48, 50]
}

df2 = pd.DataFrame(total_comments)
df2
```

```python
# Concatenatsia a doua DataFrame df + df2
# Conactenatia dupa default, pe Axis 0 !!!
pd.concat([df,df2])
```

```python
# Conactenatia df + df2 dupa Axis 1 si inscriem in df
df = pd.concat([df,df2], axis=1)
df
```

## Stergerea Coloanei

```python
# am sters coloana 'Total Comments' si am inscris in DataFrame 'df' 
df = df.drop(['Total Comments'], axis=1)
df
```

```python
# Conactenatia df + df2 dupa Axis 1 si inscriem in df
df = pd.concat([df,df2], axis=1)
df # df dupa concatenare
```

```python
df.drop(['Language'], axis=1)
# am sters caloana 'Language' dar in originala df, ea este, pentru ca sa fie 
# stearsa si din originala stergerea trebuie inscrisa in df sau cu inplace(ex. mai jos)!!
```

```python
# Stergere cu 'inplace = True' atunci se modifica DataFrame originala
df.drop(['Language'], axis=1, inplace=True)
df 
```

## Stergerea rindurilor

```python
# Stergem 2 rinduri, primul si ultimul, cu 'inplace=False' sa nu modificam
# DataFrame originala 'df'
df.drop([0,4], inplace=False)
```
