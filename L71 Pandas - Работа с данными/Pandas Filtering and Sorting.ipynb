```python
import pandas as pd 
import numpy as np
```

```python
forum_users = {
    'User ID':np.array([1, 2, 3, 4, 5]), # Putem sa folosim masive numpay pnetru creaea DataFrame
    'Username':['andrew_b', 'ion_a', 'jane_smith', 'alex123', 'bob56'],
    'Age':[34, 53, 24, 16, 65],
    'Joined Date':pd.to_datetime(['2021-01-10', '2022-03-14', '2020-01-24', '2023-03-26', '2018-11-25']),
    'Total Posts':[150, 300, 250, 30, 90],
    'Reputation':[500, 720, 600, 300, 980]
}

# Creem DataFrames 

df = pd.DataFrame(forum_users)
df
```

## Filtering

```python
# Afisham numai 2 coloane
df[['Username', 'Age']]
```

```python
df.loc[3] 
# cu 'loc[3]' gasim rindul cu metka 3 si primim o serie
```

```python
df.loc[2:4] 
# daca indicam diapazonul primi DataFrame dar nu Series
```

```python
df.loc[2:4, ['Username','Age','Reputation']] 
# indicam si colonitsele care sa e afisheze
```

```python
df.iloc[3]
```

```python
df.iloc[2:4] 
# cu 'iloc' ultima marime din diapazon se exclude
```

```python
df.iloc[2:4, 1:4]
# 'iloc' lucreaza cu indexurile
# Aici primi 2 rinduri si 3 coloane
```

```python
df.iloc[:, 1:4]
# ':' inseamna toate valoarele
```

```python
df[df['Age']>=25] # Filtrare a coloane 'Age'
```

```python
df['Age']>=25
```

```python
df[df['Total Posts'] >=200]
```

```python
df[df['Reputation'] == 500]
```

```python
df[(df['Total Posts'] >=200) & (df['Age']>=25)]
```

```python
df[(df['Total Posts'] >=400) | (df['Age']<=35)]
```

```python
df['Reputation'].isin([600,500]) # Gasim acelea care au sau 600 sau 500
```

```python
df[df['Reputation'].isin([600,500])] # Gasim acelea care au sau 600 sau 500
```

```python
df[df['Reputation'].isin(range(200,900))] # indicam diapazon
```

```python
df[df['Total Posts'].isin(range(50,250))]
```

```python
date_range = pd.date_range(start= '2018-01-01', end='2022-01-01')
date_range
```

```python
df[df['Joined Date'].isin(date_range)] 
# filtram dupa diapazonul datei, creat mai sus
```
